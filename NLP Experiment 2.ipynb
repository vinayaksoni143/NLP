```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stopwords=list(set(stopwords.words('english')))
stopwords
```

```python
s='Hello darkness my old friend, I have come to talk to you again'
tokenize_words=word_tokenize(s)
tokenize_words
```

```python
words_filtered=[w for w in tokenize_words if w not in stopwords]
words_filtered
```

```python
filtered_sentence=[]

for w in tokenize_words:
  if w not in stopwords:
    filtered_sentence.append(w)

filtered_sentence
```

```python
#Removing punctuations

tokenizer=nltk.RegexpTokenizer(r'\w+') #[a-zA-Z][^\s]*\b
tokens=tokenizer.tokenize(s)
tokens
```

```python
#stemming
from nltk.stem import PorterStemmer

from nltk.tokenize import word_tokenize,sent_tokenize

ps=PorterStemmer()

example_words=['python','pythoner','pythoning','pythoned','knowledge','lokhande']

for w in example_words:
  print(ps.stem(w))
```

```python
#POS tagging-classifieer
import nltk
from nltk import word_tokenize
text='why oh why tell me why not me?'
tokens=word_tokenize(text)
tags=nltk.pos_tag(tokens,tagset='universal')
tags
```

```python
# 2(b)
from nltk.corpus import names 
import random 
 
male_names = names.words('male.txt')
female_names = names.words('female.txt')
 
labeled_male_names = [(str(name), 'male') for name in male_names]
labeled_female_names = [(str(name), 'female') for name in female_names]

labeled_all_names = labeled_male_names + labeled_female_names

random.shuffle(labeled_all_names)
 
print("First 15 random labeled combined names:")
print (labeled_all_names[:15])
 
```
