##### Convert word into numerical representation - word embedding
##### Techniques for word embedding : <br> 
1. Bag Of Words <br> - Based of frequency of word in sentence. Completely misses the context. Each word is represented by its frequency count, which acts as its vector. Aka one hot vector representation. <br>
2. Document-Term Matrix <br> - Before using this representation you should remove the stopwords. 


```python
import nltk
from nltk.corpus import brown
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = brown

chapter_texts = []
for i in range(5):
    chapter_text = ' '.join(corpus.words(corpus.fileids()[i]))
    chapter_texts.append(chapter_text)

vectorizer = TfidfVectorizer(stop_words='english', smooth_idf=True)
corpus_matrix = vectorizer.fit_transform(chapter_texts)

print(corpus_matrix.toarray())
```
