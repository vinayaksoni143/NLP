```python
import nltk
from nltk.stem import WordNetLemmatizer

word = "teaching"
pos_tag = "v"

lemmatizer = WordNetLemmatizer()

lemma_word = lemmatizer.lemmatize(word, pos=pos_tag)

inflected_words=[lemma_word]
prefix=['s','ing','er','ed']
for i in prefix:
  inflected_words.append(lemma_word+i)

print(f"Original word: {word}")
print(f"Lemma form: {lemma_word}")
print(f"Inflected forms: {inflected_words}")
```
