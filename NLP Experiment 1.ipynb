```python
#1(a)Write a Python NLTK program to split the text sentence/paragraph from genesis corpus and display it into a list of words. 
#Remove the Punctuation and Stopwords from the given text and perform Stemming and POS tagging.

import nltk
from nltk.corpus import genesis
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import word_tokenize
```

```python
genesis.words()
```

```python
text = genesis.raw('english-web.txt')
text
```

```python
tokenizer = RegexpTokenizer(r'\w+')
sentenceList = tokenizer.tokenize(text)
print(sentenceList)
```

```python
stopwords = stopwords.words('english')
wordList = [x for x in sentenceList if x not in stopwords]
print(wordList)
```

```python
import string
sentenceList = [x for x in sentenceList if x not in string.punctuation]
print(sentenceList)
```

```python
from nltk.tag import pos_tag
tags = pos_tag(sentenceList)
print(tags)
```
