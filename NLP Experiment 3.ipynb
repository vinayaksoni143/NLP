```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.chunk import RegexpParser
from nltk.corpus import stopwords

text=state_union.raw('1945-Truman.txt')
sent_tokens=sent_tokenize(text)
text=sent_tokens[0]
tokens = word_tokenize(text)

# stopwords=list(set(stopwords.words('english')))
# filtered_text=[w for w in tokens if w not in stopwords]

tags=nltk.pos_tag(tokens)

chunker = RegexpParser(r'''
NP:
{<DT><NN.*><.*>*<NN.*>}
}<VB.*>{
''')

chunker.parse(tags)
```

```python
#Chinking
import nltk
from nltk.corpus import state_union
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.chunk import RegexpParser
from nltk.corpus import stopwords

text=state_union.raw('1945-Truman.txt')
sent_tokens=sent_tokenize(text)
text=sent_tokens[0]
tokens = word_tokenize(text)

# stopwords=list(set(stopwords.words('english')))
# filtered_tokens=[w for w in tokens if w not in stopwords]

tags=nltk.pos_tag(tokens)

chinking = RegexpParser(r"""
    NP:
    {<.*>+}          
    }<VBD|IN>+{      
  """)


chinking.parse(tags)
```

```python
import nltk
from nltk.corpus import conll2000
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.chunk import RegexpParser
from nltk.corpus import stopwords

text=conll2000.raw('test.txt')
sent_tokens=sent_tokenize(text)
text=sent_tokens[0]
tokens = word_tokenize(text)

# stopwords=list(set(stopwords.words('english')))
# filtered_text=[w for w in tokens if w not in stopwords]

tags=nltk.pos_tag(tokens)

chunker = RegexpParser(r'''
NP:
{<DT><NN.*><.*>*<NN.*>}
}<VB.*>{
''')

chunker.parse(tags)
```

```python
from nltk.corpus import conll2000
conll2000.chunked_sents('train.txt', chunk_types='NP')[20]

regParser = nltk.RegexpParser("")  # using no grammer

test_sents = conll2000.chunked_sents('test.txt', chunk_types=['NP'])

print(regParser.evaluate(test_sents))

regParser2 = nltk.RegexpParser(r"NP: {<[CDJNP].*>+}")

print(regParser2.evaluate(test_sents))
```
